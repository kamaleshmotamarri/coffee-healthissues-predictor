## Univariate analysis

### Subtask:
Explore the distribution of individual variables using histograms, box plots, and count plots.


**Reasoning**:
Generate histograms and box plots for numerical columns to visualize their distributions and identify outliers.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['Age', 'Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours', 'BMI', 'Heart_Rate', 'Physical_Activity_Hours']

# Histograms
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Generate count plots for categorical columns to show the frequency of each category.



```python
categorical_cols = ['Gender', 'Country', 'Sleep_Quality', 'Stress_Level', 'Health_Issues', 'Occupation', 'Smoking', 'Alcohol_Consumption']

plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_cols):
    plt.subplot(3, 3, i + 1)
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Frequency of {col}')
plt.tight_layout()
plt.show()
```

## Bivariate analysis

### Subtask:
Investigate the relationships between pairs of variables using scatter plots, bar plots, and correlation matrices.


**Reasoning**:
Create scatter plots for the specified pairs of numerical variables, bar plots for the specified relationships between categorical and numerical variables, and a heatmap for the correlation matrix of all numerical variables.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Scatter plots for numerical variables
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(x='Coffee_Intake', y='Caffeine_mg', data=df)
plt.title('Coffee Intake vs Caffeine mg')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Age', y='Heart_Rate', data=df)
plt.title('Age vs Heart Rate')

plt.subplot(1, 3, 3)
sns.scatterplot(x='BMI', y='Physical_Activity_Hours', data=df)
plt.title('BMI vs Physical Activity Hours')

plt.tight_layout()
plt.show()

# 2. Bar plots for categorical vs numerical variables
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.barplot(x='Country', y='Coffee_Intake', data=df, ci=None)
plt.title('Country vs Average Coffee Intake')
plt.xticks(rotation=90)

plt.subplot(1, 3, 2)
sns.barplot(x='Gender', y='BMI', data=df, ci=None)
plt.title('Gender vs BMI')

plt.subplot(1, 3, 3)
sns.barplot(x='Stress_Level', y='Heart_Rate', data=df, ci=None)
plt.title('Stress Level vs Heart Rate')

plt.tight_layout()
plt.show()

# 3. Correlation matrix and heatmap for numerical variables
numerical_cols = ['Age', 'Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours', 'BMI', 'Heart_Rate', 'Physical_Activity_Hours', 'Smoking', 'Alcohol_Consumption']
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

## Multivariate analysis

### Subtask:
Analyze relationships among three or more variables using techniques like pair plots or faceted plots.

## Define model and parameter grid

### Subtask:
Define the XGBoost classifier and the hyperparameter grid for RandomizedSearchCV.


**Reasoning**:
Define the XGBoost classifier and the hyperparameter grid for RandomizedSearchCV as requested in the subtask.



```python
from xgboost import XGBClassifier

# Create an instance of the XGBClassifier
xgb_clf = XGBClassifier(random_state=42)

```

```python
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Recreate the target variable y after filling NaN values if not already done
y = df['Health_Issues']

# Encode the target variable (XGBoost requires numeric labels)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Train/test split
X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Define the XGBClassifier with chosen hyperparameters
xgb_clf = XGBClassifier(
    n_estimators=300,      # number of trees
    learning_rate=0.1,     # step size shrinkage
    max_depth=5,           # tree depth
    subsample=0.8,         # row sampling
    colsample_bytree=0.8,  # feature sampling per tree
    gamma=0,               # min loss reduction
    random_state=42,
    n_jobs=-1              # use all CPU cores
)

# Fit directly on the training data
xgb_clf.fit(X_train, y_train_encoded)

# Predictions
y_pred_encoded = xgb_clf.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)
print(f"Accuracy on the test set: {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test_encoded, y_pred_encoded, target_names=label_encoder.classes_))
```
